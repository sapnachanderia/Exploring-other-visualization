```python
#Line Plots: Display trends over time or across ordered categories.
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = ("heart_disease_health_indicators_BRFSS2015 2 (1).csv")
data = pd.read_csv(df)

# Display the column names to check for available variables
print(data.columns)

# Function to plot line plots for trends over time
def plot_line_time(variable, time_column):
    plt.figure(figsize=(10, 6))
    sns.lineplot(x=time_column, y=variable, data=data, marker='o')
    plt.title(f'Trend of {variable} Over {time_column}')
    plt.xlabel(time_column)
    plt.ylabel(variable)
    plt.show()

# Function to plot line plots for trends across ordered categories
def plot_line_ordered(variable, ordered_column):
    plt.figure(figsize=(10, 6))
    sns.lineplot(x=ordered_column, y=variable, data=data, marker='o')
    plt.title(f'Trend of {variable} Across {ordered_column}')
    plt.xlabel(ordered_column)
    plt.ylabel(variable)
    plt.show()

# Example: Plotting trend of BMI over age
plot_line_time('BMI', time_column='Age')

# Example: Plotting trend of BMI across age categories
plot_line_ordered('BMI', ordered_column='Age')
```

```python
#Display Distributions of Continuous Variables:
import matplotlib.pyplot as plt
import seaborn as sns

# Define a function to plot histograms
def plot_histogram(column):
    plt.figure(figsize=(10, 6))
    sns.histplot(data[column], kde=True, bins=30)
    plt.title(f'Distribution of {column}')
    plt.xlabel(column)
    plt.ylabel('Frequency')
    plt.show()

# Plot histograms for BMI, MentHlth, and PhysHlth
plot_histogram('BMI')
plot_histogram('MentHlth')
plot_histogram('PhysHlth')
```

```python
#Heatmaps and Correlation Plots: Explore Variable Relationships
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Compute the correlation matrix
correlation_matrix = data.corr()

# Plot the heatmap
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap='coolwarm', cbar=True, square=True)
plt.title('Correlation Matrix Heatmap')
plt.show()

# Optional: Filter correlation matrix to focus on specific variables
variables_of_interest = ['BMI', 'MentHlth', 'PhysHlth', 'HeartDiseaseorAttack']
filtered_correlation_matrix = correlation_matrix[variables_of_interest].loc[variables_of_interest]

# Plot the filtered heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(filtered_correlation_matrix, annot=True, fmt=".2f", cmap='coolwarm', cbar=True, square=True)
plt.title('Filtered Correlation Matrix Heatmap')
plt.show()
```

```python
# Scatter Plots: Investigate Relationships Between Variables
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


# Display the column names to check for available variables
print(data.columns)

# Function to plot scatter plots
def plot_scatter(x, y):
    plt.figure(figsize=(10, 6))
    sns.scatterplot(x=x, y=y, hue='HeartDiseaseorAttack', data=data, palette='coolwarm')
    plt.title(f'{y} vs {x}')
    plt.xlabel(x)
    plt.ylabel(y)
    plt.legend(title='Heart Disease')
    plt.show()

# Scatter plot for BMI vs Mental Health Days
plot_scatter('BMI', 'MentHlth')

# Scatter plot for BMI vs Physical Health Days
plot_scatter('BMI', 'PhysHlth')

# Scatter plot for Mental Health Days vs Physical Health Days
plot_scatter('MentHlth', 'PhysHlth')

# Scatter plot for BMI vs Heart Disease
plot_scatter('BMI', 'HeartDiseaseorAttack')

# Scatter plot for Mental Health Days vs Heart Disease
plot_scatter('MentHlth', 'HeartDiseaseorAttack')

# Scatter plot for Physical Health Days vs Heart Disease
plot_scatter('PhysHlth', 'HeartDiseaseorAttack')
```

```python
#Violin Plots: Combine aspects of box plots and density plots to show the distribution of data.
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


# Display the column names to check for available variables
print(data.columns)

# Function to plot violin plots
def plot_violin(x, y='HeartDiseaseorAttack'):
    plt.figure(figsize=(10, 6))
    sns.violinplot(x=y, y=x, data=data, palette='coolwarm', split=True)
    plt.title(f'Distribution of {x} by {y}')
    plt.xlabel(y)
    plt.ylabel(x)
    plt.show()

# Violin plot for BMI
plot_violin('BMI')

# Violin plot for Mental Health Days
plot_violin('MentHlth')

# Violin plot for Physical Health Days
plot_violin('PhysHlth')
```

```python
#Pair Plots: Plot pairwise relationships across multiple variables.
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


# Display the column names to check for available variables
print(data.columns)

# Selecting a subset of variables for the pair plot
variables = ['BMI', 'MentHlth', 'PhysHlth', 'HeartDiseaseorAttack']

# Creating the pair plot
sns.pairplot(data[variables], hue='HeartDiseaseorAttack', palette='coolwarm')
plt.show()
```
